```python
def calculate_tax(row):
    salary = row['Salary_LPA']
    location = row['company_location']

    # base tax
    if salary <= 12:
        tax = 0.10
    elif salary <= 30:
        tax = 0.20
    else:
        tax = 0.30

    # metro extra tax
    if location in ['Mumbai', 'Delhi', 'Bangalore']:
        tax += 0.02

    return tax
```
