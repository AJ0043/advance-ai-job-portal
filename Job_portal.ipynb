```python
min_exp = india_df['experience_years'].min()
max_exp = india_df['experience_years'].max()

india_df['Salary_Rupees'] = (
    (india_df['experience_years'] - min_exp) / (max_exp - min_exp)
) * (50 - 10) + 10

# LPA → Rupees
india_df['Salary_Rupees'] = india_df['Salary_Rupees'] * 100000
```

```python
def calculate_tax(row):
    salary = row['Salary_LPA']
    location = row['company_location']

    # base tax
    if salary <= 12:
        tax = 0.10
    elif salary <= 30:
        tax = 0.20
    else:
        tax = 0.30

    # metro extra tax
    if location in ['Mumbai', 'Delhi', 'Bangalore']:
        tax += 0.02

    return tax
```

```python
skills_map = {
    'Data Scientist': 'Python, ML, Statistics, Pandas',
    'AI Research Scientist': 'Deep Learning, NLP, PyTorch',
    'AI Software Engineer': 'Python, APIs, ML Deployment',
    'NLP Engineer': 'NLP, Text Processing, Transformers',
    'Research Scientist': 'Research, Python, Data Analysis',
    'Head of AI': 'Leadership, AI Strategy, ML',
    'Autonomous Systems Engineer': 'Robotics, Computer Vision, AI'
}
```

```python
india_df['skills'] = india_df['job_title'].map(skills_map)
```

```python
def assign_skills(title):
    title = str(title).lower()

    if 'data scientist' in title:
        return 'Python, ML, Statistics, Pandas'
    elif 'ai research' in title:
        return 'Deep Learning, NLP, PyTorch'
    elif 'nlp' in title:
        return 'NLP, Transformers, Text Processing'
    elif 'software engineer' in title:
        return 'Python, APIs, Deployment'
    elif 'research' in title:
        return 'Research, Python, Analysis'
    elif 'head' in title:
        return 'Leadership, Strategy, AI'
    else:
        return 'General Tech Skills'

india_df['skills'] = india_df['job_title'].apply(assign_skills)
```

```python
india_df
```

```python
def add_description(skill):
    skill = str(skill).lower()

    if 'python' in skill:
        return """• Strong knowledge of Python programming (Mandatory)
• Bachelor's degree in Computer Science or related field
• 2–5 years of hands-on experience in development or data roles
• Experience in data analysis and machine learning
• Problem-solving and analytical thinking skills
• Ability to work with real-world datasets
• Good communication and teamwork abilities"""

    elif 'nlp' in skill:
        return """• Strong knowledge of Python (Mandatory)
• Bachelor's/Master’s degree in AI, Data Science or related field
• 3–6 years of experience in NLP projects
• Experience with text processing and language models
• Hands-on work with transformers and deep learning
• Strong analytical and problem-solving skills"""

    elif 'deep learning' in skill:
        return """• Strong Python programming skills (Mandatory)
• Degree in Computer Science, AI or related domain
• 4–8 years of experience in deep learning
• Experience with TensorFlow or PyTorch
• Knowledge of neural networks and model optimization
• Ability to handle large-scale data and models"""

    else:
        return """• Strong knowledge of Python (Mandatory)
• Bachelor’s degree in relevant technical field
• 2–4 years of professional experience
• Good understanding of technical concepts
• Problem-solving and teamwork skills
• Willingness to learn and adapt new technologies"""
```

```python
india_df['description'] = india_df['skills'].apply(add_description)
```
